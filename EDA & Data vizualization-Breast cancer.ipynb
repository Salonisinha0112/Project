```python
# encoding the target column 
label_encode = LabelEncoder()

labels = label_encode.fit_transform(breast_cancer_data['diagnosis'])

breast_cancer_data['target'] = labels

breast_cancer_data.drop(columns='diagnosis', axis=1, inplace=True)
```

### Data Visualization

```python
#if we have categorical column we can use countplot 
```

```python
sns.countplot(data=breast_cancer_data, x='target')
```

```python
#imbalance in the class as there is more label in 0 and less in 1
```

```python


# assume breast_cancer_data is a pandas dataframe
column_names = breast_cancer_data.columns.tolist()
print(column_names)
```

```python
#if we want columns in vertical form
for col in breast_cancer_data.columns:
    print(col)
```

```python
#univeriate ..coz we analysis one column at a time
for col in breast_cancer_data.columns:
    sns.displot(breast_cancer_data[col])
```

```python


# Loop over each column in the dataset
for col in breast_cancer_data.columns:

    # Create a distribution plot for the column
    sns.distplot(breast_cancer_data[col], kde=True, hist=False)

    # Add a vertical line for the mean value of the column
    plt.axvline(breast_cancer_data[col].mean(), color='r')

    # Set the plot title to the column name
    plt.title(col)

    # Display the plot
    plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(breast_cancer_data.columns), figsize=(10,50))

for i, col in enumerate(breast_cancer_data.columns):
    axs[i].boxplot(breast_cancer_data[col])
    axs[i].set_title(col)
    axs[i].set_ylabel(col)
```

```python
for column in breast_cancer_data:
    plt.figure()
    breast_cancer_data.boxplot([column])
```

```python
##As we can see here that most of the skewed features have Outliers
```

### Correlation Matrix

```python
correlation_matrix = breast_cancer_data.corr()
```

```python
# constructing a heat map to visualize the correlation matrix
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, cbar=True, fmt='.1f', annot=True, cmap='Blues')
plt.savefig('Correlation Heat map')
```

**Inference from EDA & Data Visualization:** 
1. No missing Values
2. All are continuous numerical values except for Target column
3. Mean is slightly more than the median for most of the features. So it is right skewed.
4. Slight imbalance in the dataset (Benign(0) cases are more than Malignant(1) cases
5. Mean of most features are clearly larger for Malignant cases compared to the benign cases (Groupby)
6. Most of the features have Outliers
7. Correlation Matrix reveal that most of the features are highly correlated. So we can remove certain features during Feature Selection
